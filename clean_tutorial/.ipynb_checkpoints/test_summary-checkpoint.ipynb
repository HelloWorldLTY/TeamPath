```python
summary_prompts = ["Provide a concise pathological summary of the tissue shown in this histopathology image, highlighting any abnormal cellular or structural features in one paragraph.",

"Based on the visual characteristics in this image, summarize the likely histological diagnosis and key indicators leading to it in one paragraph.",

"Describe the main histopathological patterns visible in this image and summarize what they suggest about the tissue state in one paragraph.",

"Summarize the key morphological findings in this histopathology image, including any signs of malignancy, inflammation, or necrosis in one paragraph.",

"Generate a pathology report-style summary based solely on this histological section, mentioning tissue type, grade, and diagnostic clues in one paragraph.",

"Briefly summarize the clinical implications of the abnormalities visible in this histopathology image in one paragraph.",

"From this histopathology image, extract and summarize the most diagnostically relevant features in one paragraph.",

"Identify and summarize any histopathological hallmarks (e.g., mitotic figures, glandular formation, stromal invasion) present in the image in one paragraph.",

"Write a summary suitable for a pathology trainee explaining what this histopathology image represents and why in one paragraph.",

"Provide an expert-level summary of the pathological findings in this histopathology image, including your confidence in the assessment in one paragraph."
]

from PIL import PngImagePlugin
LARGE_ENOUGH_NUMBER = 100
PngImagePlugin.MAX_TEXT_CHUNK = LARGE_ENOUGH_NUMBER * (1024**2)

summary_info = pd.read_csv("summary_testing_data.csv")

summary_info

# !nvidia-smi

import os
import torch
os.environ['HF_HOME'] = '/home/tl688/scratch/'
os.environ['HF_HUB_CACHE'] = '/home/tl688/scratch/'
os.environ['HF_TOKEN'] = 'your token'
```

```python
import math
import numpy as np
import torch
import torchvision.transforms as T
from decord import VideoReader, cpu
from PIL import Image
from torchvision.transforms.functional import InterpolationMode
from transformers import AutoModel, AutoTokenizer,AutoConfig
```

```python
from transformers import Qwen2_5_VLForConditionalGeneration, AutoTokenizer, AutoProcessor
from qwen_vl_utils import process_vision_info

# # default: Load the model on the available device(s)
# model = Qwen2_5_VLForConditionalGeneration.from_pretrained(
#     "/home/tl688/scratch/saves/pathor1-7b_summary/full/sft/checkpoint-790/", torch_dtype="auto", device_map="auto"
# )

# We recommend enabling flash_attention_2 for better acceleration and memory saving, especially in multi-image and video scenarios.
model = Qwen2_5_VLForConditionalGeneration.from_pretrained(
    "/home/tl688/scratch/saves/pathor1-7b_summary_5w/full/sft/checkpoint-500/",
    attn_implementation="flash_attention_2",
    device_map="auto",
)

# default processer
processor = AutoProcessor.from_pretrained("/home/tl688/scratch/saves/pathor1-7b_summary_5w/full/sft/")
```

```python
out_result = []
np.random.seed(2024)
with torch.cuda.amp.autocast():
    for index in tqdm(summary_info.index):
        row = summary_info.loc[index]
        quest = summary_prompts[0]
        wsi_path = summary_info.loc[index]['wsi_id']
        pos = eval(summary_info.loc[index]['position'])
        location = pos[0] +'_' + pos[1]
        wsi_path = wsi_path + '_' + location+'.png'
        messages = [
            {
                "role": "user",
                "content": [
                    {
                        "type": "image",
                        "image": f"./figure_caption_summary/"+wsi_path,
                    },
                    {"type": "text", "text": summary_prompts[0]},
                ],
            }
        ]

        # Preparation for inference
        text = processor.apply_chat_template(
            messages, tokenize=False, add_generation_prompt=True
        )
        image_inputs, video_inputs = process_vision_info(messages)
        inputs = processor(
            text=[text],
            images=image_inputs,
            videos=video_inputs,
            padding=True,
            return_tensors="pt",
        )
        inputs = inputs.to("cuda")

        # Inference: Generation of the output
        generated_ids = model.generate(**inputs, max_new_tokens=1024)
        generated_ids_trimmed = [
            out_ids[len(in_ids) :] for in_ids, out_ids in zip(inputs.input_ids, generated_ids)
        ]
        response = processor.batch_decode(
            generated_ids_trimmed, skip_special_tokens=True, clean_up_tokenization_spaces=False
        )
        out_result.append(response)
```

```python
import pickle 
with open('./summary_out/pathor1sft_5wtrain_full_oneprompt_ckpt500_output.pkl', 'wb') as file:
    pickle.dump(out_result, file)
```
