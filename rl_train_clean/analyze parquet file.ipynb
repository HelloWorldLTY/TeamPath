```python
import os

def find_all_folders(path):
    """
    Finds all folders, including subfolders, under the given path.

    Args:
        path (str): The starting path to search for folders.

    Returns:
        list: A list of absolute paths to all found folders.
    """
    all_folders = []
    for root, dirs, files in os.walk(path):
        for d in dirs:
            folder_path = os.path.join(root, d)
            if 'actor' in folder_path and 'merged_hf_model' not in folder_path:
                all_folders.append(folder_path)
    return all_folders
```
